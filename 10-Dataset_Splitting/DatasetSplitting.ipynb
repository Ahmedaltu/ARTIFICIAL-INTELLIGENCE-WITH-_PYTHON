```python
import pandas as pd
from sklearn import linear_model, metrics
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Admission_Predict.csv',skiprows=0,delimiter=",")
print(df.head())
```

Of the many possible variables (columns) we pick the CGPA column to be our independent variable and Chance of Admit to be our dependent variable. So we would like to investigate it CGPA can be used to predict the chance of a person to be admitted to university. In python we slice the columns on pandas data frame in usual fashion as

```python
X = df[['CGPA']]
y = df[['Chance of Admit ']]
```

```python
#Now it is time to do the splitting into training and testing portions with 80/20 ratio. This can be done in one line as
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
plt.scatter(X_train,y_train)
plt.scatter(X_test,y_test,color="red")
plt.legend(["train","test"])
plt.xlabel("CGPA")
plt.ylabel("Chance of Admit")
plt.title("Dataset splitting")
plt.show()
```

```python
print(X.head())
print(X_train.head())
print(X_test.head())
```

```python
lm = linear_model.LinearRegression()
model = lm.fit(X_train, y_train)
plt.scatter(X,y)
plt.plot(X_test,lm.predict(X_test),color="red")
plt.title("Prediction")
plt.show()
```

```python
print("R2=",lm.score(X_test,y_test)) # using linear model
print("R2=",metrics.r2_score(y_test,lm.predict(X_test)))# using sklearn
```
