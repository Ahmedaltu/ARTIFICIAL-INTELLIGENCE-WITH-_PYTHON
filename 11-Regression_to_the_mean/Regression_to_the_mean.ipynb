```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
rep= (500, 1000, 2000, 5000, 10000, 15000, 20000, 50000, 100000)
for i in rep:
  df = pd.DataFrame(np.random.randint(1,7,size=(i,2)),columns = [1,2])
  df['sum'] = df.sum(axis=1)
  s=df['sum']
  h,h2 = np.histogram(s,range(2,14))
  plt.bar(h2[:-1],h/i)
  plt.title(i)
  plt.show()
```
