```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs, make_moons
from sklearn.cluster import KMeans, SpectralClustering
```

```python
X, y_true = make_blobs(n_samples=300, centers=4,cluster_std=0.60, random_state=10)
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
#Using the kmean algorithm
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
print(y_kmeans)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans)
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', marker="+")
plt.show()
```

```python
labels = KMeans(6, random_state=0).fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.show()
```

```python
#Another limitation of kmeans is that it does not work so well with complicated geometries such as below
X, y = make_moons(200, noise=.05, random_state=0)
labels = KMeans(2, random_state=0).fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.show()
```

```python
#A solution is to apply spectral clustering algorithm as follows.
model = SpectralClustering(n_clusters=2, affinity='nearest_neighbors',
                           assign_labels='kmeans')
labels = model.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.show()
```
